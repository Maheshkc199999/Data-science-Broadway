```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("C:/Users/HP\Downloads/Salary Data.csv",usecols = ['Years of Experience','Salary'])
dataset = dataset.dropna()     
dataset
```

```python
X = dataset.iloc[:, 0:-1].values
X
```

```python
Y = dataset.iloc[:, 1:2].values
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.1,random_state = 1)
```

```python
print(X_train)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
exp = float(input("Enter years of experience = "))
exp = np.array([[exp]])
predicted_salary=model.predict(exp)[0][0]
print("Your predicted_salary = ",predicted_salary)
```

```python
predicted_salary = model.predict(X_test)
predicted_salary
```
